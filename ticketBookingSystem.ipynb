```python
import time 

print("Welcome to mini bookMyShow")
print("Aaj lag rhi hai blockbuster movie : 'Python ki Kahani' \n")

available_seats = list(range(1,51))
TICKET_PRICE = 200

while True:
    time.sleep(1)

    print("\n" +  "="*40)
    print("main menu : ")
    choice = input("1. View Available Seats \n2. Book Ticket  \n3. Exit \nEnter your choice (1/2/3): ")

    if choice == '1':
        print("\nAvailable seats: ", available_seats)

    elif choice == '2':
        if len(available_seats) == 0:
            print("Sorry, no seats available.")    
            continue
        try:
            kitni_seats = int(input("Kitni tickets book karni hai? (e.g , 2) :"))

            if kitni_seats > len(available_seats):
                print(f"sirf {len(available_seats)} seats available hai, aap {kitni_seats} tickets book kar sakte hain.")
                continue

            user_ki_seats = []

            print(f"Kripya apni {kitni_seats} fav seat number batayein : ")
            for i  in range(kitni_seats):
                seat_number = int(input(f"Seat {i+1} : "))
                user_ki_seats.append(seat_number)

            sab_sahi_hain = True 
            for seat in user_ki_seats:
                if seat not in available_seats:
                    sab_sahi_hain = False
                    print(f"Sorry, seat number {seat} is not available.")
                    break   


            if sab_sahi_hain:
                for seat in user_ki_seats:
                    available_seats.remove(seat)
                total_price = kitni_seats * TICKET_PRICE

                print("\n Booking successful!")
                print(f"Congratulations! Your seats {user_ki_seats} ")
                print(f"Total Amount Paid : Rs. {total_price}")
                print(f"Congratulations! Your seats {user_ki_seats} have been booked. Total price: Rs. {total_price}")    
        except ValueError:
            print("Invalid input. Please enter a valid number. (1,2,3,...)")

    elif choice == '3':    
        print("Thank you for using mini bookMyShow. Enjoy the movie!")
        break       
    else:
        print("Invalid choice. Please enter 1, 2, or 3.")     

```
